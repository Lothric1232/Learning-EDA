Numerical analysis

```python
numerical_cols=["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans"
                ,"Bank Deposits","Checking Accounts","Saving Accounts","Foreign Currency Account",
                "Business Lending"
                ]

plt.figure(figsize=(10,6))
for i,col in enumerate(df[numerical_cols]):
    plt.subplot(len(numerical_cols),1,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.xticks(rotation=45)
    plt.show()
```

```python
correlation_matrix=df[numerical_cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix,annot=True,cmap="crest",fmt=".2f")
plt.title("correlation_Matrix")
plt.show()
```

Insights
1)Strong Positive Correlations:
Bank Deposits and Checking Accounts have a very strong correlation of 0.84. This suggests that individuals with higher bank deposits tend to also have higher balances in their checking accounts.
Bank Loans and Foreign Currency Account also show a moderate correlation of 0.42, indicating that those who have loans are somewhat likely to have foreign currency accounts.

2)Moderate Positive Correlations:
There is a moderate correlation between Estimated Income and Superannuation Savings (0.37) as well as with Bank Loans (0.33). This implies that higher estimated income is associated with higher superannuation savings and bank loans.
Saving Accounts also have a moderate correlation with Bank Deposits (0.75) and Bank Loans (0.27).


3)Weak to Moderate Correlations:
Credit Card Balance has weaker correlations with most variables, peaking at 0.38 with Bank Deposits, indicating that while there are some links, they are not strong.
Business Lending shows a moderate correlation with Bank Loans (0.42) and a weaker association with Estimated Income (0.33).


4)Assets vs. Liabilities:
It's noticeable that bank deposits and checking accounts (both asset classes) have significantly higher correlations with each other compared to their correlations with liabilities such as loans and credit cards.
This could imply a financial behavior where individuals with more savings tend to simultaneously take out fewer loans.
